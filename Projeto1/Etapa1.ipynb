### Correlação

```python
matrizCorrelacao = df.corr()
plt.figure(figsize=(30, 10))
sns.heatmap(matrizCorrelacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

Como é possível observar pela matriz de correlação o dispositivo que mais tem correlação de dados, tanto diretamente como inversamente proporcional é o dispositivo 20. Mas tem como destaque a sua correlação inversa com o dispositivo 41.

Também, cabe ressalatar que o dispositivo 22 tem uma significante correlação inversa com o dispositivo 24.

### Histogramas

```python
num_cols = 3
num_rows = len(df.columns) // num_cols
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(4*num_cols, 2*num_rows))
for i, column in enumerate(df.columns):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    ax.hist(df[column], bins=50, color='blue', edgecolor='black')
    ax.set_title(f'Histograma de {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequência')
plt.tight_layout()
plt.show()
```

Como é possível observar todos os histogramas segue a distribuição gaussiana, portanto, pode-se confirmar os cálculos efetuados na correlação.

### Detectando anomalias nos dados

```python
from sklearn.decomposition import PCA

X = df.drop('falha', axis = 1).values #input
y = df['falha'].values #target

pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
```

Plotagem simplificada em duas dimensões separando os dados em duas classes sendo elas dados com falha e dados sem falha.

OBS: usou o drop para tirar o Tutor, pois não usa ele para detectar as anomalias


```python
from sklearn.ensemble import IsolationForest

#setar semente do gerador aleatório então da mesmo resultado
clf = IsolationForest(random_state=42)
ano_y = clf.fit_predict(X)

plt.scatter(X_pca[:,0], X_pca[:,1], c=ano_y, cmap='viridis')
```

Plotagem de dimenção reduzida dos dados normais e dados anormais. E é importante identificar a anomalias porque a qualidade do treinamento da IA depende diretamente da qualidade dos dados.

OBS: Sem o contamination a floresta isolada tenta achar o grupo mais diferente dos demais, se você sabe a porcentagem de anomalia baseado nos equipamentos utilizados é só setar o contamination.

```python
import numpy as np

a, b = np.unique(ano_y, return_counts = True)

print(a, b)
```

Importando os dados sem anomalia

```python
ano_df = df
ano_df['anomalia'] = ano_y

# Remover as linhas que são anomalias
ano_df = ano_df[ano_df['anomalia'] != -1]

#Removendo coluna de anomalia
ano_df = ano_df.drop('anomalia', axis = 1)

# Exportar para CSV
ano_df.to_csv('dados_sem_anomalias.csv', index=False)

ano_df.head()
```
